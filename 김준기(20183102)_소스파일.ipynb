```python
from bs4 import BeautifulSoup
import requests

keyword = input('원하는 키워드 입력: ')

keyword_url = requests.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query='+ 
keyword)

soup = BeautifulSoup(keyword_url.content, 'html.parser')

result = soup.find_all(attrs = {'class':'news_tit'})

for i in result:
    print(i.get_text())
    print(i.attrs['href'])
```
