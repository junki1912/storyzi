***실습 18번 ***
* 목적지, 열차 종류 및 입석/좌석 여부에 따라 기차요금을 계산하는 함수를 각각 작성하시오(단, 목적지는 서울, 대전, 대구, 울산, 부산 열차의 종류는 KTX, 새마을호, 무궁화호, 통일호로 하시오)
* 음료의 종류 및 크기를 선택하면 지불금액을 계산하는 함수를 각각 작성하시오.(단, 음료는 종류는 아메리카노, 카페라떼, 카푸치노의 크기는 스몰, 레귤러, 라지로 하시오)

```python
def traindestination(dstn):
    if dstn == '서울':
        cost_a = 10000
    elif dstn == '대전':
        cost_a = 8000
    elif dstn == '대구':
        cost_a = 6000
    elif dstn == '울산':
        cost_a = 4000
    elif dstn == '부산':
        cost_a = 2000
    return cost_a

def trainkinds(kind):
    if kind == 'KTX':
      cost_b = 5000
    elif kind == '새마을호':
      cost_b = 4000
    elif kind == '무궁화호':
      cost_b = 3000
    elif kind == '통일호':
      cost_b = 2000
    return cost_b

def seat(s):
    if s == '입석':
        cost_c = 0
    elif s == '좌석':
        cost_C = 5000
    return cost_c

def payment(cost_a, cost_b, cost_c):
    p = cost_a + cost_b + cost_c
    return print('지불하실 총 금액은 {:,}입니다.'.format(p))

dstn = input('1. 목적지 입력:\n서울 / 대전 / 대구 / 울산 / 부산\n>> ')
cost_a = traindestination(dstn)

kind = input('2. 열차종류 입력:\nKTX / 새마을호 / 무궁화호 / 통일호\n>> ')
cost_b = trainkinds(kind)

s = input('3. 좌석종류 입력:\n입석 / 좌석\n>> ')
cost_c = seat(s)

payment(cost_a, cost_b, cost_c)
```

```python
def kinds(kind):
    if kind == '카푸치노':
        cost_a = 3700
    elif kind == '카페라떼':
        cost_a = 3500
    elif kind == '아메리카노':
        cost_a = 3000
    return cost_a 

def size(sz):
    if sz == '라지':
        cost_b = 700
    elif sz == '레귤러':
        cost_b = 600
    elif sz == '스몰':
        cost_b = 500
    return cost_b

def price(cost_a, cost_b):
    price_sum = cost_a + cost_b
    return print('음료의 가격은 {:,}원 입니다.'.format(price_sum))

kind = input('1. 음료의 종류를 선택:\n카푸치노 / 카페라떼 / 아메리카노\n>> ') 
cost_a = kinds(kind)

sz = input('2. 음료의 사이즈를 선택:\n라지 / 레귤러 / 스몰\n>> ')
cost_b = size(sz)

price(cost_a, cost_b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0, 10.0) # (0.0, 5.0, 생략) - 3번째 인수가 생략되면 기본적으로 50개의 데이터가 생성됨
y1 = np.sin(np.pi * x1)
x2 = np.linspace(0.0, 5.0)
y2 = np.sin(np.pi * x2)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (20, 5))
fig.suptitle('2 subplots', fontsize = 20)

ax1.plot(x1, y1, '-', marker = 'o', color = 'limegreen')
ax1.set_ylabel('Y1-axis')

ax2.plot(x2, y2, ':', marker = '^', color = 'dodgerblue')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y2-axis')

plt.show()
```

```python
y = np.random.normal(loc = 0.5, scale = 0.4, size = 1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean()) # mean: 수학적으로 정의되는 평균값으로 산술평균, 기하평균, 조화평균 등이 있으며 특정하지 않으면 산술평균을 의미, 일상에서는 average라는 용어 사용
plt.yscale('symlog')
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.10, right = 0.95, hspace = 0.25, wspace = 0.35)

plt.show()
```
